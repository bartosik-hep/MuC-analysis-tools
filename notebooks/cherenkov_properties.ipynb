### Threshold momentum vs refractive index

```python
# y: Cherenkov threshold refractive index x: momentum [MeV]; 0: mass;
f1 = R.TF1('f1', '1/(x/sqrt([0]*[0]+x*x))', 0.0, 500.0)
f1.SetParameter(0, 0.511) # electron
```

```python
refr = [
    ('Aerogel', 1+0.0026),
    ('C4F10', 1+0.0014),
    ('CF4', 1+0.0005),
    ('CO_{2}', 1+0.00041),
    ('Air', 1+0.000273),
]
C = R.TCanvas('canvas', '', 1200, 600)
R.gStyle.SetOptStat(0)
C.SetLeftMargin(0.1)
C.SetGrid(1,1)
# Axis histogram
h_axis = R.TH1F('h_axis', ';e^{#pm} momentum [MeV];Refractive index', 100, 0, 20)
h_axis.SetAxisRange(1.0, 1.003, 'Y')
h_axis.Draw('AXIS')
h_axis.Draw('AXIG same')
f1.SetLineColor(2)
f1.SetLineWidth(2)
f1.SetNpx(500)
f1.Draw('same')
C.RedrawAxis()
C.Draw()
```
